```python
toyo4= toyo1.drop(toyo.index[[80,221,960]],axis=0)
```

```python
m4= smf.ols("Price~Age+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight", data = toyo4).fit()
m4.summary()
```

```python
finalmodel = smf.ols("Price~Age+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight", data = toyo4).fit()
finalmodel.summary()
```

```python
finalmodel_pred = finalmodel.predict(toyo4)
```

```python
plt.scatter(toyo4["Price"],finalmodel_pred,c='r');plt.xlabel("Observed values");plt.ylabel("Predicted values")
```

```python
plt.scatter(finalmodel_pred, finalmodel.resid_pearson,c='r');plt.axhline(y=0,color='blue');plt.xlabel("Fitted values");plt.ylabel("Residuals")
```

```python
plt.hist(finalmodel.resid_pearson) 
```

```python
import pylab
import scipy.stats as st
st.probplot(finalmodel.resid_pearson, dist='norm',plot=pylab)
```

```python
from sklearn.model_selection import train_test_split

train_data,test_Data= train_test_split(toyo1,test_size=0.3)
```

```python
finalmodel1 = smf.ols("Price~Age+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight", data = train_data).fit()
finalmodel1.summary()
```

```python
finalmodel_pred = finalmodel1.predict(train_data)
finalmodel_pred
```

```python
finalmodel_res = train_data["Price"]-finalmodel_pred
finalmodel_res
```

```python
finalmodel_rmse = np.sqrt(np.mean(finalmodel_res*finalmodel_res))
finalmodel_rmse
```

```python
finalmodel_testpred = finalmodel1.predict(test_Data)
finalmodel_testpred
```

```python
finalmodel_testres= test_Data["Price"]-finalmodel_testpred
finalmodel_testres
```

```python
finalmodel_testrmse = np.sqrt(np.mean(finalmodel_testres*finalmodel_testres))
finalmodel_testrmse
```
